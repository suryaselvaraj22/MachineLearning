```python
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
#Set Grid Style 
#sns.set_style('whitegrid')
```

```python
#Read Dataset

df = pd.read_csv('movies.csv')
```

```python
#Check out the data
df.head()
```

```python
#Data Cleaning

#Check for missing data

for col in df.columns:
    pct_null = np.mean(df[col].isnull())
    print('{} : {}%'.format(col, pct_null))
```

```python
#Not much missing data - but drop the missing values

df = df.dropna()
```

```python
#Check datatypes of all columns

df.dtypes
```

```python
#Remove decimals for more readability by converting them to int 

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
#If we want to keep missing values intact 
#Convert pandas series to numpy array to handle missing data 

col = np.array(df['budget'], np.int64)
df['budget'] = col
```

```python
#Similarly converting 'gross' column 

col = np.array(df['gross'], np.int64)
df['gross'] = col
```

```python
#Check datatypes now 

df.dtypes
```

# Commonly used Regex in Python:
^: Check if a string begins with a specific value
$: Check if a string ends with a specific value
.: A period matches any single characters, except for new lines
*: A star matches 0 or more occurrences of the preceding patterns
[…]: A bracket matches any single character in brackets.
[^…]: Matches any single character not in brackets
{…}: Matches exactly the number of occurrences
(…): Used to group sub-patterns
+: A plus sign matches one or more occurrences
|: Matches either values
\: Escape special characters. Put a slash before the character that you are unsure whether it has special meaning or not

```python
# Check for correctness of data 

#Released column has release date and Year column has released year 
#In some cases, they do not match -> extract year from released as a new column yearcorrect

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Check data now 
df
```

```python
#Released column has location and Location column has released location
#In some cases, they do not match -> extract location from released as a new column locationcorrect

df['locationcorrect'] = df['released'].str.extract(pat = '.*\((.*)\).*').astype(str)
```

```python
df
```

```python
# Rearrange data 
# Sort data based on decreasing order of gross column 

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

### Correlation Analysis

Making some educated guesses

The following might have high correlation to gross earnings: 
* Budget
* Company 
* Star
* Director

```python
# Scatter plot can help look at this - budget vs gross 

# For better readability
#df = df.sort_values(by=['gross'], inplace=False, ascending=False)

# Scatter plot
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Seems to be correlated but visualizations are better in seaborn 
# Regression Plot will be better for interpreting correlation

sns.regplot(x=df['budget'], y=df['gross'], data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# There's clearly some correlation -> now we can measure how much they are correlated? 
```

```python
# corr() method shows the correlation matrix for all numeric data using Pearson's method as default
# df.corr(method='pearson') => 'pearson' , 'kendall', 'spearman'

df.corr() 
```

```python
# Budget and Votes are the two features highly correlated to gross 
```

```python
# Visualizing the correlation matrix using a heat map for better understanding 

corrmat = df.corr()
sns.heatmap(corrmat, cmap='coolwarm', annot=True)
plt.title('Correlation of numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
# To attain correlation of all columns (including non-numeric ones), numerize all columns 

df_numerized = df 

for col in df_numerized:
    if(df_numerized[col].dtype == object): 
        #convert to category 
        df_numerized[col] = df_numerized[col].astype('category')
        #assign numeric codes
        df_numerized[col] = df_numerized[col].cat.codes
df_numerized
```

```python
# Now we can get the correlation matrix of this numerized dataframe 

allcorrmat = df_numerized.corr()
sns.heatmap(allcorrmat, cmap='coolwarm', annot=True)
plt.title('Correlation of all features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
# It looks like Budget and Votes are highly correlated to Gross Earnings 
# However, Star and Director are not really that correlated to Gross Earnings 
```

```python
# We can also unstack to view correlation between pairs easily 

allcorrmat = df_numerized.corr()
allcorrpairs = allcorrmat.unstack()
allcorrpairs
```

```python
# Sort to see correlation pairs 
sortedpairs = allcorrpairs.sort_values()
sortedpairs
```

```python
highcorrpairs = sortedpairs[sortedpairs > 0.5]
highcorrpairs
```

```python
# It looks like Budget and Votes are highly correlated to Gross Earnings 
# However, Star and Director are not really that correlated to Gross Earnings 
# One of my educated guess was right and I found a new feature which has a high correlation to Gross Earnings!
```
